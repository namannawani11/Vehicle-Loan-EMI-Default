```python
# precision
precision_lr = TP/(TP+FP)
precision_lr
```

```python
# recall(true positive rate)
recall_lr = TP/(TP+FN)
recall_lr
```

```python
# specificity (True negative rate)
specificity_lr = TN/(TN+FP)
specificity_lr
```

```python
f1_score_lr = 2*((precision_lr*recall_lr)/(precision_lr+recall_lr))
f1_score_lr
```

```python
accuracy_lr = (TN+TP)/(TN+FP+FN+TP)
accuracy_lr
```
